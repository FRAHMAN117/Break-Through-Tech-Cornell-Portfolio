CLEANING OUTLIERS

```python
import scipy.stats as stats
```

```python
#as each column is part of a social data, eahc data is highly varible. Thats why instead of removing outliers, I winsorize which will set any outlier to the the top and botton 1%
df_store= df.copy() #stores the current dataframe

# as these are technically labels and not numerical values, they are excluded from winsorizing.
columns2= ['country','year', 'lifeL'] 
df_win= df_store.drop(columns= columns2).columns 


for i in df_win:
    df[i + 'win'] = stats.mstats.winsorize(df[i], limits=[0.01, 0.01])


#checks for the distribution
df.describe()
```

```python
#add the winzoried dataframe to the main one
df.drop(columns= df_win, inplace= True)
df.describe()
```

```python
print(df.columns.tolist())
```

```python
#renames the winsorized columns
df.rename(columns={'log_GDPwin': 'log_GDP', 'socialSwin':'socialS', 'life_Expwin':'life_Exp'}, inplace=True)
df.rename(columns={'freedomwin': 'freedom','generositywin': 'generosity', 'corruptionwin': 'corruption'}, inplace= True)
df.rename(columns={'positivewin': 'positive','negativewin': 'negative', 'gov_confidencewin': 'gov_confidence'}, inplace= True)
df.rename(columns={'dem_Qualitywin': 'dem_Quality','dev_Qualitywin': 'dev_Quality','stdwin' : 'std'}, inplace= True)
df.rename(columns={'stdMeanwin': 'stdMean', 'gini_indexwin': 'gini_index', 'gini_Avgwin': 'gini_Avg'}, inplace= True)
df.rename(columns={'gini_householdIncwin': 'gini_householdInc'}, inplace= True)
df.head()
```

ONE HOT ENCODING

```python
#to hot encode the values, we need to see what values are appropriate and eligible for hot encoding. 
df.dtypes
```

```python
#as the country is the only objecy and not a numerical value or feature, we encode this into specific countries which will be later used to see which countries are the happiest or saddest
to_encode= list(df.select_dtypes(include= ['object']).columns)
to_encode
```

```python
df[to_encode].nunique() #shows the total number of values we need to hot encode
```

```python
#takes the top 20 most common  countries represented in the dataframe

top_20_cn= list(df['country'].value_counts().head(20).index) 
top_20_cn
```

```python
#here is where the values get turned into binary columns
for value in top_20_cn:
    df['country_' + value]= np.where(df['country']== value,1,0) #the name is also changed to give them a different representation
df.drop(columns='country', inplace= True) #remove the country column as it has been hot encoded
```

```python
df.head()
```

FEATURE SELECTION/ FEATURE ENGINEERING

```python
#drop the features not used. as std and stdmean is a part of the target variable (it comes form lifeL), we remove this to reduce data leakage
feature_drop=['std', 'stdMean']
df.drop(columns= feature_drop, inplace= True)

print(df.columns.tolist())
```

FEATURE SCALINING AND SPLITING DATA

```python
#also removes the target vairable ro reduce data leakage
y= df['lifeL']
X= df.drop(columns= 'lifeL', axis= 1)

print("Number of examples: " + str(X.shape[0]))
print("\nNumber of Features:" + str(X.shape[1]))
print(str(list(X.columns)))
```

```python
#libraries for linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

STANDARIZATION

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1234) 
#divides the data into training and testing set (70% training and 30% testing set)
```

```python
#used AI for standarization as to scale regression problem. This problem cannot be scaled by normalization
from sklearn.preprocessing import StandardScaler
noncon=[]
scalar= StandardScaler()
for i in X_train.columns:
    if not i.startswith('country_'):
        noncon.append(i)
X_train.loc[:, noncon]= scalar.fit_transform(X_train[noncon])
X_test.loc[:, noncon]= scalar.transform(X_test[noncon])
```

LINEAR REGRESSION MODEL

```python
#create linear regression model
model= LinearRegression()

#train model by training the set
model.fit(X_train, y_train)


y_predict= model.predict(X_test) #this is the variable used for prediction

#prints the first 5 predicted and actual values
print(f"Predictions {y_predict[:5]}") 
print(f"actual results{y_test.head(5)}")
```

```python
#creates a scatter plot with a bets fit line to measure how much does the predicted target variable differes form actual target variable
plt.title("Life ladder compared to features")
plt.xlabel("Actual Life Ladder")
plt.ylabel("predicted life ladder")

sns.regplot(x=y_test, y= y_predict)
```

```python
#examines model parameters

#intercept 
print('Model Summary:\n')
print('Intercept')
print('alpha= ', model.intercept_)

features= X.columns
#weights
print('\n weights:')
i=0
for w in model.coef_:
    print('w_',i+1,'= ', w, ' [ weight of ', features[i],']') #measure the weight for each hot encoded value, menaing every country
    i += 1
```

```python
#mean squared error by finding the difference between squaring the predicted value from actual target variablr
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, y_predict)))

print(' R^2 =   %.2f'
      % r2_score(y_test, y_predict))
```

RANDOM FOREST 

```python
#libraries for Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#def train_test_RFR(X_train, X_test, y_train, y_test, n_estimators, min_samples_leaf, max_depth, criterion='squared_error', random_state= 13):

#trains RF model

rmodel= RandomForestRegressor(random_state=13) #random state is the number of ranodm variables
rmodel.fit(X_train, y_train)

#shows the predicted and actual target variables
y_rpredict= rmodel.predict(X_test)
print(f"Predictions \n{y_rpredict[:5]}\n")
print(f"actual results\n{y_test.head(5)}\n")

#Calculates the mean squared error, absolute error, rmse, and r2 score
#initializes the variables similar to linear regression
mar= mean_absolute_error(y_test,y_rpredict)
msr=mean_squared_error(y_test,y_rpredict)
r2=r2_score(y_test,y_rpredict)
rmse= np.sqrt(msr)
    
#prints variables
print(f"Mean absolute error: {mar}\n")
print(f" Mean Squared error: {msr}\n")


#shows the model formance before tuning
print('\nModel Performance\n\nRMSE =   %.2f'
    % rmse)

print(' R^2 =   %.2f'
    % r2)
    
    
#return mar, msr, r2, rmse, rmodel
    
```

```python
#this section is debugged by AI. These are all the tesing parameters that will be used to find the best parameter.
param_grid={
    'n_estimators': [100,200,300],
    'max_depth': [10,20,30],
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,4]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#GridSearchSV model is initialized the test values are set to itself and the rmodel is being estimated for the best hyperparameters
rCV= GridSearchCV(estimator= rmodel, param_grid= param_grid, cv=3, scoring= 'neg_mean_squared_error') 
```

```python
rCV.fit(X_train, y_train) #model is fit similar to a normal model
```

```python
yCV_rpredict= rCV.predict(X_test) #model is predicted similar to a normal model
```

```python
#starts grid search by gathering the mean and absulte squared error for each node.
print("grid search...\n")
marCV= mean_absolute_error(y_test,yCV_rpredict)
msrCV=mean_squared_error(y_test,yCV_rpredict)
r2CV=r2_score(y_test,yCV_rpredict)
rmseCV= np.sqrt(msrCV)

print("Best Parameters\n")
print(rCV.best_params_)
    
#prints variables after calculating error
print(f"Mean absolute error: {marCV}\n")
print(f" Mean Squared error: {msrCV}\n")

#prints model performance after tuning
print('\nModel Performance\n\nRMSE =   %.2f'
      % rmseCV)
   
print(' R^2 =   %.2f'
      % r2CV)
```

IDENTIFY THE MOST IMPORTANT FEATURES

```python
#this section is imporant as it is used to find the more important featuresthat play the most part in happiness in this data. 
importantFea_model = RandomForestRegressor(min_samples_leaf = 1, max_depth = 30, criterion = 'squared_error', random_state=13)
importantFea_model.fit(X_train, y_train)

fi = importantFea_model.feature_importances_
fi[:5] #gets the 5 most importanr features
```

```python
names_sorted = [x for _,x in sorted(zip(fi, X_train.columns.values), reverse = True)] #sorts the imporant features in order.
```

```python
#graph the feature value to give visualized in feature importance
fig, ax = plt.subplots()

ax.bar(np.arange(10), sorted(importantFea_model.feature_importances_, reverse=True)[:10], width = 0.35)
ax.set_xticks(np.arange(10))
ax.set_xticklabels(names_sorted[:10], rotation = 90)
plt.title('Feature importance from RF')
ax.set_xlabel('Feature Columns')
ax.set_ylabel('Normalized importance')
```
